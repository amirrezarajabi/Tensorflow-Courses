```python
import tensorflow
import numpy as np
import matplotlib.pyplot as plt
```

```python
from tensorflow.keras import layers, Model
from tensorflow.keras.applications.inception_v3 import InceptionV3
weights_file = "./datasets-models/inception_v3_weights_tf_dim_ordering_tf_kernels_notop.h5"
```

```python
pre_trained_model = InceptionV3(
    input_shape=(150, 150, 3),
    include_top=False,
    weights=None
)
pre_trained_model.load_weights(weights_file)
```

```python
for layer in pre_trained_model.layers:
    layer.trainable = False

pre_trained_model.summary()
```

```python
last_layer = pre_trained_model.get_layer("mixed7")
last_output = last_layer.output
last_layer.output_shape
```

```python
x = tensorflow.keras.layers.Flatten()(last_output)
x = tensorflow.keras.layers.Dense(1024, activation=tensorflow.nn.relu)(x)
x = tensorflow.keras.layers.Dropout(0.2)(x)
x = tensorflow.keras.layers.Dense(1, activation=tensorflow.nn.sigmoid)(x)
model = Model(pre_trained_model.input, x)
model.compile(
    optimizer=tensorflow.keras.optimizers.RMSprop(lr=0.001),
    loss=tensorflow.keras.losses.binary_crossentropy,
    metrics=["accuracy"]
)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
training_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)
test_datagen = ImageDataGenerator(rescale=1./255)

train_genetator = training_datagen.flow_from_directory(
    "./datasets-models/cats_and_dogs_filtered/train/",
    batch_size=20,
    class_mode="binary",
    target_size=(150, 150),
)

test_generator = test_datagen.flow_from_directory(
    "./datasets-models/cats_and_dogs_filtered/validation/",
    batch_size=20,
    class_mode="binary",
    target_size=(150, 150),
)
```

```python
history = model.fit_generator(
    train_genetator,
    validation_data=test_generator,
    steps_per_epoch=100,
    epochs=20,
    validation_steps=50,
    verbose=2
)
```

```python
acc, loss, val_acc, val_loss = history.history["accuracy"], history.history["loss"], history.history["val_accuracy"], history.history["val_loss"]
epochs = range(len(acc))

plt.plot(epochs, acc)
plt.plot(epochs, val_acc)
plt.show()
```

```python
plt.plot(epochs, loss)
plt.plot(epochs, val_loss)
plt.show()
```
